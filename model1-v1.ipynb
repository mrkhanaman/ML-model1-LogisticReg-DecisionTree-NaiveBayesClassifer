# Machine Learning Model 1: Classification Model using Logisting Regression, Decision Tree Classifer and Naive Bayes Classifier.

```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
#reading data
df=pd.read_csv("D:/Aman/Social_Network_Ads.csv")
df.head()
```

```python
df['Gender'].replace(['Male', 'Female'],[0, 1], inplace=True)
df.head()
```

```python
df.info()
```

```python
#input and output values
X=df.iloc[:, [1,2,3,4]].values
Y=df.iloc[:, 5].values
```

```python
#splitting the data
from sklearn.model_selection import train_test_split
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X, Y, test_size = 0.20, random_state = 0)
```

# Model 1: Logistic Regression

```python
#Model1
from sklearn.linear_model import LogisticRegression
m1 = LogisticRegression(random_state = 0)
m1.fit(X_Train, Y_Train)
```

```python
#predicting using model
Y_Pred1 = m1.predict(X_Test)
```

```python
from sklearn import metrics
print(metrics.accuracy_score(Y_Test,Y_Pred1))
```

```python
#Evaluating the model
from sklearn.metrics import confusion_matrix
labels = np.unique(Y_Test)
cm = confusion_matrix(Y_Test, Y_Pred1, labels=labels)
pd.DataFrame(cm, index=labels, columns=labels)
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_Test, Y_Pred1))
```

# Model 2: Decision Tree Classifier

```python
#Model2
from sklearn.tree import DecisionTreeClassifier
m2= DecisionTreeClassifier()
m2.fit(X_Train, Y_Train)
```

```python
#predition and evaluating the model
Y_Pred2=m2.predict(X_Test)
print(metrics.accuracy_score(Y_Test,Y_Pred2))
```

```python
cm2 = confusion_matrix(Y_Test, Y_Pred2, labels=labels)
pd.DataFrame(cm2, index=labels, columns=labels)
```

```python
print(classification_report(Y_Test, Y_Pred2))
```

# Model 3: Naive Bayes Classifer

```python
#model3
from sklearn.naive_bayes import GaussianNB
m3 = GaussianNB()
m3.fit(X_Train, Y_Train)
```

```python
#predition and evaluating the model
Y_Pred3=m3.predict(X_Test)
print(metrics.accuracy_score(Y_Test,Y_Pred3))
```

# Evaluation: Confusion Matrix

```python
cm3 = confusion_matrix(Y_Test, Y_Pred3, labels=labels)
pd.DataFrame(cm3, index=labels, columns=labels)
```

```python
print(classification_report(Y_Test, Y_Pred3))
```
